```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/home/jashnoor/Documents/ML Internship/task 1/Titanic-Dataset.csv')
```

```python
print("\nSummary Statistics:")
print(df.describe())
print("\nData Info:")
print(df.info())
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
num_cols = ['Age', 'Fare', 'SibSp', 'Parch']
df[num_cols].hist(figsize=(10, 8), bins=15, edgecolor='black')
plt.suptitle('Histograms of Numeric Features')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
for i, col in enumerate(num_cols):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(y=df[col], color='lightblue')
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.pairplot(df[['Survived', 'Age', 'Fare', 'Pclass', 'Sex']].dropna())
plt.suptitle('Pairplot of Selected Features', y=1.02)
plt.show()
```

```python
for col in num_cols:
    skew = df[col].skew()
    print(f"Skewness in {col}: {skew:.2f}")
```
